#**Data Toolkit Assignment**

1. What is NumPy, and why is it widely used in Python ?
  - NumPy (Numerical Python) is a library for numerical operations. It offers powerful array objects, broadcasting, and tools for integrating C/C++ and Fortran code. It's widely used due to its efficiency, speed, and ease of use in handling large datasets and mathematical operations.
  ---
2.  How does broadcasting work in NumPy ?
  - Broadcasting allows NumPy to perform arithmetic operations on arrays of different shapes by automatically expanding one to match the other’s shape, without copying data.
  ---
3. What is a Pandas DataFrame ?
  - A Pandas DataFrame is a 2D, mutable, and labeled data structure with columns that can be of different data types, similar to a table in SQL or Excel.
  ---
4. Explain the use of the groupby() method in Pandas?
  - groupby() is used to group data based on one or more keys, then apply aggregation functions (like sum, mean) to each group independently.
  ---
5. Why is Seaborn preferred for statistical visualizations?
  - Seaborn is built on Matplotlib but provides a higher-level interface for drawing attractive and informative statistical graphics with minimal code.
  ---
6. What are the differences between NumPy arrays and Python lists?
  - NumPy arrays are more efficient and faster.

  - Support element-wise operations.

  - Homogeneous types.

  - Lists are general-purpose, support multiple data types, and are less efficient for numerical tasks.
  ---
7. What is a heatmap, and when should it be used?
  - A heatmap is a color-coded matrix representing data. It's used to visualize correlation matrices or to show intensity variations in data.
  ---
8. What does the term “vectorized operation” mean in NumPy?
  - It refers to applying operations on entire arrays rather than element by element, which is faster and more efficient.
   ---
9. How does Matplotlib differ from Plotly?
  - Matplotlib is static and used for basic plotting.

  - Plotly is interactive and web-based with rich functionalities.
  ---
10. What is the significance of hierarchical indexing in Pandas?
  - It allows multiple index levels in rows or columns, enabling complex data organization and retrieval.
  ---
11. What is the role of Seaborn’s pairplot() function?
  - It plots pairwise relationships in a dataset and is useful for exploratory data analysis (EDA).
  ---
12. What is the purpose of the describe() function in Pandas?
  - Returns summary statistics like mean, count, std, min, max, etc., for numerical columns.
  ---
13. Why is handling missing data important in Pandas?
  - Missing data can bias results or break calculations. Handling it ensures data integrity and accurate analysis.
  ---
14. What are the benefits of using Plotly for data visualization?
  - Interactive, web-friendly, highly customizable, supports 3D plots and dashboards.
  ---

15. How does NumPy handle multidimensional arrays?
  - It uses ndarray objects which support N-dimensional data with efficient storage and operations.
  ---
16. What is the role of Bokeh in data visualization?
  - Bokeh is a Python library for interactive and scalable visualizations in web browsers.
  ---
17. Explain the difference between apply() and map() in Pandas?
  - map() is used on Series only and works element-wise.

  - apply() can be used on both DataFrames and Series and can apply functions across axis.
  ---
18. What are some advanced features of NumPy?
  - Broadcasting

  - Masking

  - Structured arrays

  - Memory-mapped files

  - Linear algebra and Fourier transforms
  ---
19. How does Pandas simplify time series analysis?
  - With date-time indexing, resampling, shifting, time zone support, and rolling window calculations.
  ---
20. What is the role of a pivot table in Pandas?
  - It summarizes data by grouping and aggregating based on column values, similar to Excel pivot tables.
  ---
21. Why is NumPy’s array slicing faster than Python’s list slicing?
  - NumPy arrays use contiguous memory blocks and vectorized operations, making slicing a view operation instead of copying.
  ---
22. What are some common use cases for Seaborn?
  - Distribution plots

  - Correlation analysis

  - Time series plots

  - Regression analysis

  - Categorical plots

```python
#Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df['B'].mean())
```

```python
#Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y)
```

```python
#How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'a':[1,2,3],'b':[4,5,6],'c':[7,8,9]})
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Generate a bar plot using Plotly
import plotly.express as px
df = px.data.tips()
fig = px.bar(df, x='day', y='total_bill', color='sex')
fig.show()
```

```python
#Create a DataFrame and add a new column based on an existing column
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]})
df['B'] = df['A'] * 2
print(df)
```

```python
#Write a program to perform element-wise multiplication of two NumPy arrays
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
#Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3]
y1 = [2, 3, 4]
y2 = [4, 3, 2]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
#Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold
import pandas as pd
df = pd.DataFrame({'A': [10, 20, 30]})
filtered = df[df['A'] > 15]
print(filtered)
```

```python
#Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt
data = [1, 2, 2, 3, 3, 3, 4]
sns.histplot(data, kde=True)
plt.show()
```

```python
#Perform matrix multiplication using NumPy
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```
